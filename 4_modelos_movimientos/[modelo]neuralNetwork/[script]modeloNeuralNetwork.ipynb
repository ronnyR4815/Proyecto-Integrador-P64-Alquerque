```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from joblib import dump

tf.random.set_seed(77)

# Cargar los datos
data = pd.read_csv('mov_train.csv')

# Separar dataset y etiquetas
X_train = data.drop(columns=['validez'])
y_train = data['validez']

# Crear el modelo
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compilar el modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entrenar el modelo
model.fit(X_train, y_train, epochs=500, batch_size=20, validation_split=0.2)
```

```python
dump(model, '[modelo]NeuralNetwork.joblib')
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
from joblib import load

# Cargar los datos
data = pd.read_csv('mov_test.csv')

# Separar dataset y etiquetas
X_test = data.drop(columns=['validez'])
y_test = data['validez']

# Cargar el modelo entrenado
model = load('0.joblib')

# Realizar predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Loss: {loss}')
print(f'Accuracy: {accuracy}')
```
